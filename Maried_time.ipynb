## 擷取文章內結婚日期並轉換成阿拉伯數字

```python
dict = {u'零':0, u'一':1, u'二':2, u'三':3, u'四':4, u'五':5, u'六':6, u'七':7, u'八':8, u'九':9, u'十':10, u'百':100,u'千':1000, u'万':10000,u'0':0, u'1':1, u'2':2, u'3':3, u'4':4, u'5':5, u'6':6, u'7':7, u'8':8, u'9':9,u'壹':1, u'贰':2, u'叁':3, u'肆':4, u'伍':5, u'陆':6, u'柒':7, u'捌':8, u'玖':9, u'拾':10, u'佰':100, u'仟':1000,u'萬':10000,u'亿':100000000}
bicdict = {u'十':10,u'百':100,u'千':1000,u'万':10000,u'拾':10,u'佰':100,u'仟':1000,u'萬':10000,u'亿':100000000}
intdict={u'0':0, u'1':1, u'2':2, u'3':3, u'4':4, u'5':5, u'6':6, u'7':7, u'8':8, u'9':9}

def convertword_digi(a, encoding="utf-8"):
    if isinstance(a, str):
        a = a.decode(encoding)
    count = 0
    result = 0
    tmp = 1      #為了單一個'十'設定
    control = 0
    aa=""
    while count < len(a):
        tmpChr = a[count]
        #print tmpChr
        if tmpChr not in dict:  #剔除意外字
            control = 1
            count += 1
        elif tmpChr in intdict: #阿拉伯數字計算邏輯與中文不同
            aa=aa+str(a[count])
            count += 1                    
        else:
            tmpNum = dict.get(tmpChr, None)  #所有十進位單位
            if tmpChr in bicdict:
                result = result + tmp * tmpNum
                tmp = 0
            else:                #個位數
                tmp = tmpNum
            count += 1
    if result != 0:
        result = result + tmp   #轉換後數字
    else:
        if len(str(aa))==1:
            result="0"+aa
        else:
            result = aa         #原阿拉伯數字
    return result
```
