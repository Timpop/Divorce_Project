## 各地區Tf-idf挑出之關鍵字(最下)

```python
import glob,codecs,re,jieba
from sklearn import feature_extraction  
from sklearn.feature_extraction.text import TfidfTransformer  
from sklearn.feature_extraction.text import CountVectorizer
filetent=[]
worddic={}
wordlist=[]
filelist=[]
for filename in glob.glob('E:\\iii_DataSet\\Case\\sinicut_location\\*.txt'):       
    with codecs.open( filename ,'r' ,'utf-8') as f :
        mm=re.search(r'E:\\iii_DataSet\\Case\\sinicut_location\\(\w*).txt',filename)
        casename=mm.group(1)
        filelist.append(casename)
        content = f.read() 
        content2=re.sub(u'[\s ，、]','',content)
        #print content2
        f.close()         
        m=re.split(u'\d*',content2)
        casetent=','.join(m)
        filetent.append(casetent)


vectorizer=CountVectorizer(min_df=1)#该类会将文本中的词语转换为词频矩阵，矩阵元素a[i][j] 表示j
transformer=TfidfTransformer()#该类会统计每个词语的tf-idf权值
tfidf=transformer.fit_transform(vectorizer.fit_transform(filetent)) #corpus:欲處理文本
#第一个fit_transform是计算tf-idf，第二个fit_transform是将文本转为词频矩阵
word=vectorizer.get_feature_names()#获取词袋模型中的所有词语
stopword = vectorizer.get_stop_words()
weight=tfidf.toarray()
for i in xrange(len(weight)):   #len(weight)=4    weight-->清單類別的特徵矩陣    
    ww=open('E:\\iii_DataSet\\Case\\sinicut_location\\weight\\'+filelist[i]+\
                    '_weight.txt','w')
    print u"-------这里输出第",i+1,u"类文本的词语tf-idf权重------"
    for j in xrange(len(word)):
        if weight[i][j]*100 > 0:
            print word[j],"%.5f" % (weight[i][j]*100) 
            wordlist.append(word[j]+":"+"%.5f" % (weight[i][j]*100))            
            ww.write(word[j].encode('utf8')+":"+"%.5f" % (weight[i][j]*100)+"\n")
print "-------------------------------------------------------------------"
ww.close()
```
