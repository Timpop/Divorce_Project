```python
import glob,codecs,re,jieba
from sklearn import feature_extraction  
from sklearn.feature_extraction.text import TfidfTransformer  
from sklearn.feature_extraction.text import CountVectorizer
filetent=[]
for filename in glob.glob('E:\\iii_DataSet\\Case\\sinicut_annual\\weight\\*.txt'):       
    with codecs.open( filename ,'r','utf8') as f :
        mm=re.search(r'E:\\iii_DataSet\\Case\\sinicut_annual\\weight\\(\w*).txt',filename)
        casename=mm.group(1)
        content=f.read()
        #print content
        f.close()
        filetent.append(content)
vectorizer=CountVectorizer(min_df=1)#该类会将文本中的词语转换为词频矩阵，矩阵元素a[i][j] 表示j
transformer=TfidfTransformer()
tfidf=transformer.fit_transform(vectorizer.fit_transform(filetent)) #corpus:欲處理文本
word=vectorizer.get_feature_names()
weight=tfidf.toarray()
for i in xrange(len(weight)):   #len(weight)=4    weight-->清單類別的特徵矩陣    
    ww=open('E:\\iii_DataSet\\Case\\sinicut_annual\\weight\\'+filelist[i]+\
                    '_weight.txt','w')
    print u"-------这里输出第",i+1,u"类文本的词语tf-idf权重------"
    for j in xrange(len(word)):
        if weight[i][j]*100 > 0:
            print word[j],"%.5f" % (weight[i][j]*100)       
            ww.write(word[j].encode('utf8')+":"+"%.5f" % (weight[i][j]*100)+"\n")
print "-------------------------------------------------------------------"
ww.close()
```
